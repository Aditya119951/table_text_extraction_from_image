```python
import cv2
import pytesseract
import pandas as pd
import numpy as np
import re

# Path to the image file
image_path = '/kaggle/input/image2/image2.png'

# Read the image
image = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Use pytesseract to perform OCR on the grayscale image
extracted_text = pytesseract.image_to_string(gray)

# Remove "|" from the extracted text
extracted_text = extracted_text.replace("|", "").replace("[","").replace("]","").replace("{","").replace("}","")

# Split the text into rows
rows = extracted_text.strip().split('\n')

# Remove the empty spaces
cleaned_list = [item for item in rows if item.strip()]
```

```python
cleaned_list
```

```python
# Split the first element (header) by whitespace to get column names
columns = cleaned_list[0].split()

# Remove the first element (header) from the list
cleaned_list = cleaned_list[1:]

# Combine non-numeric words with space between them
try:
    cleaned_list = [re.sub(r'(\D) (\D)', r'\1\2', item) for item in cleaned_list]
except:
    pass

# Convert the list into a DataFrame
df = pd.DataFrame([x.split() for x in cleaned_list], columns=columns)

# Print the DataFrame
df
```
